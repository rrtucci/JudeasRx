# Plotting dot files with graphviz
Bnets are stored as bif or dot files. bif files include the structure and the TPMs of the bnet, whereas dot files only 
include structure info. 

bif and dot files can be created with the BayesNet and BifTool classes of JudeasRx. 

graphviz plots dot files, as shown below.

bif files for more complicated bnets can be found at

http://www.bnlearn.com/bnrepository/

```python
import os
import sys
os.chdir('../')
sys.path.insert(0,os.getcwd())
print(os.getcwd())
```

```python
import graphviz as gv
from IPython.display import display, Image


def draw(path):
    s = gv.Source.from_file(path)
    
    # using display(s) will draw the graph but will not
    # embed it permanently in the notebook. To embed it
    # permanently, must generate temporary image file and use Image(). 
    # display(s)
    
    x = s.render("jupyter_notebooks/tempo", format='png',view=False)
    display(Image(x))
```

```python
draw("examples_cbnets/asia.dot")
# TB-HRA= Tuberculosis High Risk Area
# dyspnea= shortness of breath
```

```python
draw("examples_cbnets/WetGrass.dot")
```
